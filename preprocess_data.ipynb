```python
import pandas as pd

# Load JSON file into a DataFrame
faqs = pd.read_json("data/cleaned_faqs.json")

# Display the first few rows
print(faqs.head())
```

```python
# Extract unique words inside {{}} across all rows
pattern = r"\{\{(.*?)\}\}"  # Regular expression to find text inside {{}}

# Flatten all matches into a single list
matches = faqs["question"].str.findall(pattern).explode()

# Get unique values
unique_values = matches.dropna().unique()

# Display result
print(unique_values)
```

```python
# Extract unique words inside {{}} across all rows
pattern = r"\{\{(.*?)\}\}"  # Regular expression to find text inside {{}}

# Flatten all matches into a single list
matches2 = faqs["answer"].str.findall(pattern).explode()

# Get unique values
unique_values2 = matches2.dropna().unique()

# Display result
print(unique_values2)
```
